# Função para localizar habilidade e competência dado um número do ITEM


function_compara:
a função pega o dt_habilidade e procura a **informação da habilidade** e informação da **competencia do item** de acordo com as variaveis valor_habilidade e valor_area.



* dt_competencia = dataset com todas as competencias de acordo com area
* dt_habilidade = dataset com todas as habilidades e seus valores de dificuldade
* sg_area = area da prova




Função **get_item** = retorna um item especifico de uma determinada prova

* dt_itens_provas = dataset com todos os itens da prova
* co_item = codigo do item a ser buscado
* co_prova = codigo da prova

* pra saber a informação da habilidade eu preciso do C0_HABILIDADE e SG_AREA do item

```python
def get_habilidade(dt_itens_provas, co_item, co_prova, dt_habilidade, sg_area, dt_competencia):

  df_item = dt_itens_provas[(dt_itens_provas['CO_ITEM'] == co_item) & (dt_itens_provas['CO_PROVA'] == co_prova)] #primeiro busco o item X na prova Y a habilidade do item
  valor_habilidade = df_item.iat[0,4]  # pego a habilidade do item
  df_habilidade_aux = dt_habilidade[(dt_habilidade['CO_HABILIDADE'] == valor_habilidade) & (dt_habilidade['SG_AREA'] == sg_area)] # Com a habilidade do item consigo retornar p dataframe da matriz habilidade desse item
  df_competencia_aux = dt_competencia[(dt_competencia['SG_AREA'] == df_habilidade_aux.iat[0,0]) & (dt_competencia['CO_COMPETENCIA'] == df_habilidade_aux.iat[0,1])] #pego a competencia e suas informações

  df_resultado = pd.DataFrame({
    'INF_COMPETENCIA': df_competencia_aux['INF_COMPETENCIA'],
    'INF_HABILIDADE': df_habilidade_aux['INF_HABILIDADE']
  })

  return df_resultado
  #return ([df_competencia_aux['INF_COMPETENCIA'], df_habilidade_aux['INF_HABILIDADE']])
```
