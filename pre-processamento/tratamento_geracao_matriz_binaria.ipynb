# Função para geração das matrizes de 0 e 1 de acordo com cada area

```python
def function_test(mat):
 # dic_anuladas = dict() #dicionario para verificar as questoes anuladas
  linhas_alunos = mat.shape[0] # pegar o total de linhas (registros de estudantes)
  tam_vetor_gabarito = len(mat[0][2]) #pega o total de questoes do vetor gabarito, serve para identificar tbm se é LC
  tam_respostas = len(mat[0][1])
  mat_principal = [] # matriz de retorno

  for x in range(linhas_alunos): #linha
    aux = [] #lista
    aux.append(mat[x][0]) # pega a coluna de inscricao
    vetor_gabarito = []  # Initialize vetor_gabarito here
    if(tam_vetor_gabarito != tam_respostas): #eh uma prova de LC com as questões de Espanhol e Ingles
      if(mat[x][3] == 0): # verificar se lingua escolhida da prova foi ingles
        vetor_gabarito = mat[x][2] #pegar o vetor gabarito
        vetor_gabarito = vetor_gabarito[0:5]+vetor_gabarito[10:] #concatenar somente as questoes de ingles com o resto das questoes (excluindo as de espanhol)
      else: #espanhol
        vetor_gabarito = mat[x][2] #pegar o vetor gabarito
        vetor_gabarito = vetor_gabarito[5:] #caso seja espanhol as 5 primeiras de ingles saem do filtro
      for z in range(tam_respostas): # Laço para percorrer a TX_RESPOSTA_CN e TX_GABARITO_CN
        if(vetor_gabarito[z] != 'X'): #verificar se existe questao que foi anulada (nao vai entrar na matriz binaria de erro e acerto)
          if(mat[x][1][z] == vetor_gabarito[z]): #se a resposta for igual
            aux.append(1)
          else: #verifica se a escolha de prova eh diferente quando temos um TX_RESPOSTA = 9
            aux.append(0)
        else:
          #dic_anuladas[area] = z+1
          aux.append(2)
    else:
      for z in range(tam_respostas): # Laço para percorrer a TX_RESPOSTA_CN e TX_GABARITO_CN
          if(mat[x][2][z] != 'X'):
            if(mat[x][1][z] == mat[x][2][z]): #se a resposta for igual
              aux.append(1) #adiciona 1 como certo
            else:
              aux.append(0) #errouuu
          else:
            aux.append(2)

    mat_principal.append(aux) # list of list
  column_names = ['matricula'] + ['Q' + str(i) for i in range(1, len(mat_principal[0]))]  # Create column names
  df = pd.DataFrame(data=mat_principal, columns=column_names)  # Create DataFrame
  # Remove columns with all values equal to 2
  df = df.loc[:, (df != 2).any(axis=0)]

  return df
  #verificar questoes em branco no TX_resposta
```

# Criar dataframe das matrizes de erros e acertos de acordo com cada area (Transformar em função)


```python
def criar_matriz_binaria(dt_dados):
  array = np.array(dt_dados) #Teste usando dtdados_AM_CN
  df_dados = function_test(array) #matriz auxiliar pra receber o retorno da funcao
  raw_score = df_dados[df_dados.columns[1:46]].sum(axis=1)
  df_dados['raw_score'] = raw_score
  return df_dados
```

```python
def pegar_respostas_alunos(df):
    matriz_respostas_alunos = []
    df_prova = np.array(df)
    qtde_linhas = df_prova.shape[0]
    tam_respostas = len(df_prova[0][1])

    for i in range(qtde_linhas):
        aux = []
        aux.append(df_prova[i][0])
        for j in range(tam_respostas):
            if(df_prova[i][2][j] != 'X'):
                aux.append(df_prova[i][1][j])
            else:
                aux.append(2)
        matriz_respostas_alunos.append(aux)

    matriz_respostas_alunos
    column_names = ['matricula'] + ['Q' + str(i) for i in range(1, len(matriz_respostas_alunos[0]))]  # Create column names
    df_respostas_alunos = pd.DataFrame(data=matriz_respostas_alunos, columns=column_names)  # Create DataFrame
    # Remove columns with all values equal to 2
    df_respostas_alunos = df_respostas_alunos.loc[:, (df_respostas_alunos != 2).any(axis=0)]
    df_respostas_alunos

    return df_respostas_alunos
```

```python
def somar_coluna(df, coluna):
    """
    Retorna a soma dos valores de uma coluna específica em um DataFrame.

    Args:
        df (pandas.DataFrame): O DataFrame contendo os dados.
        coluna (str): O nome da coluna a ser somada.

    Returns:
        float: A soma dos valores da coluna.
    """
    return df[coluna].sum()
```
